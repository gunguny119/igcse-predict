```python
import pandas as pd
```

```python
qp2 = pd.read_csv('component2_2016-2021.csv')
ms2 = pd.read_csv('component2_ms_2016-2021.csv')
qp46 = pd.read_csv('component46_2016-2021.csv')
```

```python
processed_qp2 = qp2[qp2['text'].apply(lambda x: x.count('?')) == 1]
```

```python
df = pd.merge(ms2, processed_qp2, how='outer', on=['year', 'month', 'component', 'question number'])
```

```python
all_df = pd.concat([df, qp46], ignore_index=True)
```

```python
all_df
```

```python
all_df.to_csv('all_2016-2021.csv', index=False)
```
